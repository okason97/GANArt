```python
class ImageDatasetWrapper():
    def __init__(self, dataset):
        self.dataset=dataset
        
    def __getitem__(self, key):
        if isinstance(key, slice):
            return [np.asarray(self.dataset[i][0]) for i in range(*key.indices(len(self.dataset)))]
        elif isinstance(key, int):
            return np.asarray(self.dataset[key][0])
        
    def __len__(self):
        return len(self.dataset)
```

```python
# images = []
shape = 256
count = 0
img_count = 0
for label in labels:
    if not os.path.exists('small-wikiart/'+label+'/'):
        os.makedirs('small-wikiart/'+label+'/')
    for image_name in listdir('wikiart/{}'.format(label)):
        img = io.imread('wikiart/{}/{}'.format(label, image_name))
        min_dim = np.argmin(img.shape[0:2])
        scale = shape/img.shape[min_dim]
        x = math.ceil(scale*img.shape[0])
        y = math.ceil(scale*img.shape[1])
        resized_img = resize(img, (x,y), anti_aliasing=True)
        center_x = int((resized_img.shape[0]-shape)/2)
        center_y = int((resized_img.shape[1]-shape)/2)
        cropped_img = resized_img[center_x:center_x+shape,center_y:center_y+shape]
        io.imsave('small-wikiart/{}/{}'.format(label, image_name), img_as_ubyte(cropped_img))
        # images.append([img, label])
        clear_output(wait=True)
        print('{}[{}]'.format(count/len(labels)*100,img_count))
        img_count+=1
    count+=1
```
